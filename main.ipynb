```python
import pandas as pd
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
air_pollution = pd.read_csv('data/air_pollution.csv')
```

```python
cities_pollution = pd.read_csv('data/cities_pollution.csv')
```

```python
world_cities = pd.read_csv('data/world_cities.csv')
```

```python
air_pollution.head()
```

```python
air_pollution = air_pollution.rename(columns={'Location': 'Country'})
air_pollution = air_pollution.rename(columns={'FactValueNumeric': 'Mortality rate of country (per 100,000)'})
air_pollution.head()
```

```python
air_pollution = air_pollution[['Country','Mortality rate of country (per 100,000)']]

air_pollution
```

```python
cities_pollution.head()
```

```python
cities_pollution = cities_pollution[['City', 'Country', 'AirQuality','WaterPollution']]
cities_pollution.head()
```

```python
world_cities = world_cities[['City', 'Country', 'population','lat', 'lng']]
world_cities = world_cities.rename(columns={'population': 'population of city'})
world_cities.head()
```

```python
gdp_2020 = pd.read_csv('data/gdp_2020.csv')
```

```python
gdp_2020.head()
```

```python
merge_1=pd.merge(cities_pollution, world_cities, on= ['Country', 'City'])
merge_1
```

```python
merge_1 = merge_1.dropna(how = 'any')
merge_1
```

```python
groupby_1 = merge_1.groupby("Country")["population of city"].sum()
groupby_1.name = 'population of country'
merge_1=pd.merge(groupby_1, merge_1, on= 'Country')

merge_1
```

```python
merge_1['proportion of population']= merge_1['population of city']/merge_1['population of country']
merge_1
```

```python
merge_2 = pd.merge(air_pollution, gdp_2020, on= 'Country')
merge_2
```

```python
merge_3 = pd.merge(merge_1, merge_2, on= 'Country')
merge_3
```

```python
merge_3['Weighted mortality rate of city (per 100,000)'] = merge_3['Mortality rate of country (per 100,000)']*merge_3['proportion of population']
merge_3
```

```python
engine=sqlalchemy.create_engine('sqlite:///data/db.sqlite')
```

```python
merge_3.to_sql('data', if_exists= 'replace', index=False, con=engine)
```
